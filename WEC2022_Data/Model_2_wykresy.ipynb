```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('model_2_df.csv')
```

```python
df.describe()
```

```python
df.columns
```

```python
col_names = ['STAY_LENGTH_D',
 'COUPON_NUMBER',
 'TIME_DEPARTURE_LOCAL_TIME',
 'FLIGHT_DISTANCE',
 'TOTAL_PRICE_PLN',
 'BOOKING_DOMESTIC_FLAG',
 'PAX_N',
 'sale_to_flight_time',
 'flight_len',
 'same_carrier',
 'is_sus_currency',
 'FLIGHT_RANGE_LONG-HAUL',
 'FLIGHT_RANGE_SHORT-HAUL',
 'VAB_SAVER',
 'PAX_TYPE_INFANT',
 'TRIP_TYPE_ONE WAY',
 'TRIP_TYPE_ROUND TRIP',
]
Y = df['purchase_time_diff']
df = df[col_names]
```

```python
X = df#.drop(columns='purchase_time_diff')
#Y = df['purchase_time_diff']
```

```python
best_hypers = {'booster': 'gbtree', 'colsample_bytree': 1.0, 'eta': 0.05, 'gamma': 1.0, 'max_depth': 14, 'min_child_weight': 1.0, 'n_estimators': 268, 'objective': 'reg:squarederror', 'subsample': 0.65}
```

```python
import xgboost as xgb
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.dtypes
```

```python
xgb_model = xgb.XGBRegressor(
    seed=42,
    use_label_encoder=False,
    gpu_id=0, **best_hypers
)
xgb_model.fit(X_train, y_train, eval_metric="rmse")
```

```python
import dalex as dx
przyklad = X.iloc[42]
xgb_exp = dx.Explainer(xgb_model, X[col_names], Y, label='XGB Pipeline')
```

```python
bd_przyklad = xgb_exp.predict_parts(przyklad,
             type = 'break_down')
bd_przyklad.result
```

```python
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

```python
bd_przyklad.plot(max_vars=30)
print('Prawdziwy wynik:', y_test.iloc[42])
```

```python
cp_przyklad = xgb_exp.predict_profile(przyklad)
cp_przyklad.result
```

```python
for col in X.columns:
    cp_przyklad.plot(variables=[col])
```
